```python
# ============================================================
# ENHANCED ADAPTIVE FACTOR-REGIME ML MODEL — NIFTY 50
# v4.0 | Up-to-date signals | Zero look-ahead on latest data
# ============================================================
# Install first (Colab):
!pip install hmmlearn yfinance xgboost lightgbm scikit-learn optuna matplotlib seaborn --quiet

import warnings
warnings.filterwarnings("ignore")
import copy
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import seaborn as sns
import optuna
import datetime
optuna.logging.set_verbosity(optuna.logging.WARNING)

from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    ConfusionMatrixDisplay, RocCurveDisplay, brier_score_loss
)
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
import lightgbm as lgb
from hmmlearn.hmm import GaussianHMM

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
print("✅ All imports successful.")

# ============================================================
# SECTION 1 — GLOBAL CONFIG
# ============================================================
START_DATE      = "2005-01-01"
END_DATE        = datetime.date.today().strftime("%Y-%m-%d")   # always fetch to today
TODAY           = datetime.date.today().strftime("%Y-%m-%d")
print(f"📅 Running model as of: {TODAY}")
TOP_K           = 10
FORWARD_MONTHS  = 6
TRAIN_RATIO     = 0.60
VAL_RATIO       = 0.20
N_REGIMES       = 3
OPTUNA_TRIALS   = 40

SLIPPAGE        = 0.001
IMPACT          = 0.0005
TARGET_VOL      = 0.14
MAX_WEIGHT      = 0.15
SECTOR_CAP      = 0.30

BUY_THRESH      = 0.62
SELL_THRESH     = 0.42
INITIAL_CAPITAL = 1_000_000   # ₹10 lakh

NIFTY50_TICKERS = [
    "ADANIENT.NS","ADANIPORTS.NS","APOLLOHOSP.NS","ASIANPAINT.NS",
    "AXISBANK.NS","BAJAJ-AUTO.NS","BAJAJFINSV.NS","BAJFINANCE.NS",
    "BHARTIARTL.NS","BPCL.NS","BRITANNIA.NS","CIPLA.NS",
    "COALINDIA.NS","DIVISLAB.NS","DRREDDY.NS","EICHERMOT.NS",
    "GRASIM.NS","HCLTECH.NS","HDFCBANK.NS","HDFCLIFE.NS",
    "HINDALCO.NS","HINDUNILVR.NS","ICICIBANK.NS","INFY.NS",
    "ITC.NS","JSWSTEEL.NS","KOTAKBANK.NS","LT.NS","MARUTI.NS",
    "NESTLEIND.NS","NTPC.NS","ONGC.NS","POWERGRID.NS",
    "RELIANCE.NS","SBIN.NS","SBILIFE.NS","SHRIRAMFIN.NS",
    "SUNPHARMA.NS","TATACONSUM.NS","TATAMOTORS.NS","TATASTEEL.NS",
    "TCS.NS","TECHM.NS","TITAN.NS","ULTRACEMCO.NS","WIPRO.NS"
]

SECTOR_MAP = {
    "ADANIENT.NS":"Energy",      "ADANIPORTS.NS":"Infra",
    "RELIANCE.NS":"Energy",      "ONGC.NS":"Energy",
    "BPCL.NS":"Energy",          "COALINDIA.NS":"Energy",
    "POWERGRID.NS":"Utilities",  "NTPC.NS":"Utilities",
    "HDFCBANK.NS":"Financials",  "ICICIBANK.NS":"Financials",
    "SBIN.NS":"Financials",      "KOTAKBANK.NS":"Financials",
    "AXISBANK.NS":"Financials",  "BAJFINANCE.NS":"Financials",
    "BAJAJFINSV.NS":"Financials","SBILIFE.NS":"Financials",
    "HDFCLIFE.NS":"Financials",  "SHRIRAMFIN.NS":"Financials",
    "INFY.NS":"IT",              "TCS.NS":"IT",
    "WIPRO.NS":"IT",             "TECHM.NS":"IT","HCLTECH.NS":"IT",
    "HINDUNILVR.NS":"FMCG",      "ITC.NS":"FMCG",
    "NESTLEIND.NS":"FMCG",       "BRITANNIA.NS":"FMCG","TATACONSUM.NS":"FMCG",
    "MARUTI.NS":"Auto",          "TATAMOTORS.NS":"Auto",
    "EICHERMOT.NS":"Auto",       "BAJAJ-AUTO.NS":"Auto",
    "ULTRACEMCO.NS":"Cement",    "GRASIM.NS":"Cement",
    "ASIANPAINT.NS":"Consumer",  "TITAN.NS":"Consumer",
    "APOLLOHOSP.NS":"Healthcare","SUNPHARMA.NS":"Pharma",
    "DRREDDY.NS":"Pharma",       "DIVISLAB.NS":"Pharma","CIPLA.NS":"Pharma",
    "JSWSTEEL.NS":"Metals",      "TATASTEEL.NS":"Metals","HINDALCO.NS":"Metals",
    "BHARTIARTL.NS":"Telecom",   "LT.NS":"Infra",
}

PALETTE = ["#2196F3","#FF9800","#4CAF50","#E91E63","#9C27B0","#00BCD4"]
print("✅ Config loaded.")

# ============================================================
# SECTION 2 — DATA DOWNLOAD
# ============================================================
def fix_df_columns(df):
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [str(col[0]) for col in df.columns]
    df = df.reset_index()
    for c in list(df.columns):
        if str(c).lower() in ("date","datetime","index"):
            df.rename(columns={c: "date"}, inplace=True)
            break
    df["date"] = pd.to_datetime(df["date"])
    return df

print("\n📥 Downloading NIFTY 50 index data...")
nifty_raw = yf.download("^NSEI", start=START_DATE, end=END_DATE,
                         interval="1mo", auto_adjust=True, progress=False)
if isinstance(nifty_raw.columns, pd.MultiIndex):
    nifty_raw.columns = nifty_raw.columns.get_level_values(0)
nifty_raw.index = pd.to_datetime(nifty_raw.index).to_period("M").to_timestamp("M")
nifty_idx = nifty_raw[["Close","High","Low","Volume"]].copy().rename(
    columns={"Close":"nifty_close","High":"nifty_high",
             "Low":"nifty_low","Volume":"nifty_vol"})
nifty_idx["nifty_ret"]  = nifty_idx["nifty_close"].pct_change()
nifty_idx["nifty_vol3"] = nifty_idx["nifty_ret"].rolling(3).std()
nifty_idx.dropna(inplace=True)
nifty_idx = fix_df_columns(nifty_idx)
print(f"   Index rows: {len(nifty_idx)}")

print("\n📥 Downloading individual stock data...")
frames, failed = [], []
for s in NIFTY50_TICKERS:
    try:
        df = yf.download(s, start=START_DATE, end=END_DATE,
                         interval="1mo", auto_adjust=True, progress=False)
        if df.empty or len(df) < 30:
            failed.append(s); continue
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)
        df.index = pd.to_datetime(df.index).to_period("M").to_timestamp("M")
        df = df[["Close","High","Low","Volume"]].rename(
            columns={"Close":"price","High":"high","Low":"low","Volume":"volume"})
        df["stock"]  = s
        df["sector"] = SECTOR_MAP.get(s, "Other")
        df = fix_df_columns(df)
        frames.append(df)
    except Exception as e:
        failed.append(s)
        print(f"   ⚠️  {s}: {e}")

stocks_raw = pd.concat(frames, ignore_index=True)
stocks_raw.columns = [str(c) for c in stocks_raw.columns]
print(f"   Stocks: {stocks_raw['stock'].nunique()}  |  Failed: {failed}")

# ============================================================
# SECTION 3 — FEATURE ENGINEERING (30+ features)
# ============================================================
print("\n🔧 Engineering features (30+ indicators)...")

stocks = stocks_raw.sort_values(["stock","date"]).copy().reset_index(drop=True)
for col in ["price","volume","high","low"]:
    stocks[col] = pd.to_numeric(stocks[col], errors="coerce")

# Returns
for n in [1,3,6,12]:
    stocks[f"ret_{n}m"] = stocks.groupby("stock")["price"].pct_change(n)

# Momentum (lag-1 to prevent look-ahead)
for n, mp in [(3,2),(6,3),(12,6)]:
    stocks[f"mom_{n}m"] = stocks.groupby("stock")["ret_1m"].transform(
        lambda x, n=n, mp=mp: x.shift(1).rolling(n, min_periods=mp).sum())

# Volatility
for n, mp in [(3,2),(6,3),(12,6)]:
    stocks[f"vol_{n}m"] = stocks.groupby("stock")["ret_1m"].transform(
        lambda x, n=n, mp=mp: x.rolling(n, min_periods=mp).std())

# Moving-average ratios
for n, mp in [(3,2),(12,6),(24,12)]:
    stocks[f"ma_{n}m"] = stocks.groupby("stock")["price"].transform(
        lambda x, n=n, mp=mp: x.rolling(n, min_periods=mp).mean())
stocks["price_to_ma3"]  = stocks["price"] / stocks["ma_3m"]
stocks["price_to_ma12"] = stocks["price"] / stocks["ma_12m"]
stocks["price_to_ma24"] = stocks["price"] / stocks["ma_24m"]
stocks["ma_cross"]      = stocks["ma_3m"] / stocks["ma_12m"]

# RSI
def compute_rsi(series, period=14):
    delta = series.diff()
    gain  = delta.clip(lower=0).rolling(period, min_periods=period//2).mean()
    loss  = (-delta.clip(upper=0)).rolling(period, min_periods=period//2).mean()
    rs    = gain / loss.replace(0, np.nan)
    return 100 - (100 / (1 + rs))

stocks["rsi_14"] = stocks.groupby("stock")["price"].transform(compute_rsi)

# Bollinger Band %
def bb_pct(series, w=20, k=2):
    mid   = series.rolling(w, min_periods=w//2).mean()
    std   = series.rolling(w, min_periods=w//2).std()
    upper = mid + k*std; lower = mid - k*std
    return (series - lower) / (upper - lower + 1e-9)

stocks["bb_pct"] = stocks.groupby("stock")["price"].transform(bb_pct)

# MACD normalised
def macd_signal(series):
    ema12 = series.ewm(span=12, adjust=False).mean()
    ema26 = series.ewm(span=26, adjust=False).mean()
    macd  = ema12 - ema26
    sig   = macd.ewm(span=9, adjust=False).mean()
    return (macd - sig) / (series.abs().rolling(12).mean() + 1e-9)

stocks["macd_norm"] = stocks.groupby("stock")["price"].transform(macd_signal)

# 52-week high proximity
stocks["hi_52w"]   = stocks.groupby("stock")["price"].transform(
    lambda x: x.rolling(12, min_periods=6).max())
stocks["prox_52w"] = stocks["price"] / stocks["hi_52w"]

# Volume momentum & ratio
stocks["vol_mom"]   = stocks.groupby("stock")["volume"].transform(
    lambda x: x.pct_change(3))
stocks["vol_ratio"] = stocks.groupby("stock")["volume"].transform(
    lambda x: x / x.rolling(6, min_periods=3).mean())

# Sector-relative momentum
stocks["sec_mom_rel"] = stocks.groupby(["date","sector"])["ret_6m"].transform(
    lambda x: x - x.median())

# Cross-sectional rank features
for col in ["mom_6m","vol_12m","rsi_14"]:
    stocks[f"{col}_rank"] = stocks.groupby("date")[col].rank(pct=True)

# Rolling Beta (3-year)
print("   Computing 3-year rolling beta …")
nifty_ret_map = nifty_idx.set_index("date")["nifty_ret"].to_dict()
stocks["nifty_ret"] = stocks["date"].map(nifty_ret_map)

def rolling_beta(group, window=36):
    sr = group["ret_1m"].values; mr = group["nifty_ret"].values
    betas = []
    for i in range(len(group)):
        if i < window-1:
            betas.append(np.nan); continue
        s_ = sr[i-window+1:i+1]; m_ = mr[i-window+1:i+1]
        mask = ~(np.isnan(s_) | np.isnan(m_))
        if mask.sum() < 10:
            betas.append(np.nan)
        else:
            cv = np.cov(s_[mask], m_[mask])
            betas.append(cv[0,1]/cv[1,1] if cv[1,1]!=0 else np.nan)
    return betas

beta_vals = []
for _, grp in stocks.groupby("stock", sort=False):
    beta_vals.extend(rolling_beta(grp))
stocks["beta_3y"] = beta_vals

# Idiosyncratic volatility
stocks["idio_vol"] = stocks["ret_1m"] - stocks["beta_3y"] * stocks["nifty_ret"]
stocks["idio_vol"] = stocks.groupby("stock")["idio_vol"].transform(
    lambda x: x.rolling(12, min_periods=6).std())

# Nifty macro features
nifty_macro     = nifty_idx.set_index("date")[["nifty_ret","nifty_vol3"]].to_dict()
stocks["mkt_ret_1m"] = stocks["date"].map(nifty_macro["nifty_ret"])
stocks["mkt_vol_3m"] = stocks["date"].map(nifty_macro["nifty_vol3"])

# ── Target: 6-month forward outperformance ─────────────────────────────────
# CHANGE: use np.where so recent rows (no future yet) get NaN target instead
# of a spurious label — prevents look-ahead on the latest months.
nifty_ret_series = nifty_idx.set_index("date")["nifty_ret"]
stocks["fwd_ret_6m"] = stocks.groupby("stock")["ret_1m"].transform(
    lambda x: x.shift(-FORWARD_MONTHS).rolling(FORWARD_MONTHS, min_periods=3).sum())

unique_dates  = stocks["date"].unique()
nifty_fwd_map = {}
for d in unique_dates:
    try:
        loc_ = nifty_ret_series.index.get_loc(d)
        end_ = loc_ + FORWARD_MONTHS
        if end_ > len(nifty_ret_series):
            nifty_fwd_map[d] = np.nan
        else:
            vals_ = nifty_ret_series.iloc[loc_:end_]
            nifty_fwd_map[d] = vals_.sum() if vals_.notna().sum() >= 3 else np.nan
    except:
        nifty_fwd_map[d] = np.nan

stocks["nifty_fwd_6m"] = stocks["date"].map(nifty_fwd_map)

# Use NaN where future is unknown (most recent FORWARD_MONTHS months)
stocks["target"] = np.where(
    stocks["fwd_ret_6m"].notna() & stocks["nifty_fwd_6m"].notna(),
    (stocks["fwd_ret_6m"] > stocks["nifty_fwd_6m"]).astype(float),
    np.nan
)

FEATURE_COLS = [
    "ret_1m","ret_3m","ret_6m","ret_12m",
    "mom_3m","mom_6m","mom_12m",
    "vol_3m","vol_6m","vol_12m",
    "price_to_ma3","price_to_ma12","price_to_ma24","ma_cross",
    "rsi_14","bb_pct","macd_norm","prox_52w",
    "vol_mom","vol_ratio","sec_mom_rel",
    "mom_6m_rank","vol_12m_rank","rsi_14_rank",
    "beta_3y","idio_vol",
    "mkt_ret_1m","mkt_vol_3m",
]

# ── Split: rows with known targets vs truly latest rows (no future yet) ────
stocks_for_model = stocks.dropna(subset=FEATURE_COLS + ["target","fwd_ret_6m"]).copy()
stocks_latest    = stocks[stocks["fwd_ret_6m"].isna()].copy()
stocks_latest    = stocks_latest.dropna(subset=FEATURE_COLS).copy()

# Winsorise based on stocks_for_model distribution only (no look-ahead)
winsor_bounds = {}
for col in FEATURE_COLS:
    lo = stocks_for_model[col].quantile(0.01)
    hi = stocks_for_model[col].quantile(0.99)
    winsor_bounds[col] = (lo, hi)
    stocks_for_model[col] = stocks_for_model[col].clip(lo, hi)
    if col in stocks_latest.columns:
        stocks_latest[col] = stocks_latest[col].clip(lo, hi)

stocks_for_model.reset_index(drop=True, inplace=True)
stocks_latest.reset_index(drop=True, inplace=True)

print(f"   Rows with targets  : {len(stocks_for_model)} (used for train/val/test)")
print(f"   Latest rows (live) : {len(stocks_latest)}  (used for BUY/HOLD/SELL signals)")
print(f"   Stocks={stocks_for_model['stock'].nunique()} | Features={len(FEATURE_COLS)}")
print(f"   Target balance: {stocks_for_model['target'].value_counts(normalize=True).round(3).to_dict()}")
if len(stocks_latest) > 0:
    print(f"   Latest signal date : {stocks_latest['date'].max().date()}")

# ============================================================
# SECTION 4 — CHRONOLOGICAL SPLIT  (uses stocks_for_model)
# ============================================================
print("\n📅 Chronological split...")
all_dates = sorted(stocks_for_model["date"].unique())
n_d       = len(all_dates)
train_end = int(n_d * TRAIN_RATIO)
val_end   = int(n_d * (TRAIN_RATIO + VAL_RATIO))

train_dates = set(all_dates[:train_end])
val_dates   = set(all_dates[train_end:val_end])
test_dates  = set(all_dates[val_end:])

train_df = stocks_for_model[stocks_for_model["date"].isin(train_dates)].copy()
val_df   = stocks_for_model[stocks_for_model["date"].isin(val_dates)].copy()
test_df  = stocks_for_model[stocks_for_model["date"].isin(test_dates)].copy()

print(f"   TRAIN: {min(train_dates).date()} → {max(train_dates).date()} ({len(train_df)} rows)")
print(f"   VAL  : {min(val_dates).date()} → {max(val_dates).date()}  ({len(val_df)} rows)")
print(f"   TEST : {min(test_dates).date()} → {max(test_dates).date()}  ({len(test_df)} rows)")

fig, ax = plt.subplots(figsize=(13, 2.5))
ax.barh(0, len(train_dates), left=0,                color="#2196F3", label="Train 60%")
ax.barh(0, len(val_dates),   left=len(train_dates),  color="#FF9800", label="Val 20%")
ax.barh(0, len(test_dates),  left=len(train_dates)+len(val_dates), color="#4CAF50", label="Test 20%")
ax.set_yticks([]); ax.set_xlabel("Monthly Steps")
ax.set_title("Chronological Train / Val / Test Split"); ax.legend()
plt.tight_layout(); plt.show()

# ============================================================
# SECTION 5 — REGIME DETECTION (Robust Fix)
# ============================================================
print("\n🔍 Detecting market regimes...")

# Use all stocks data (including latest) so regimes extend to today
stocks_all = pd.concat([stocks_for_model, stocks_latest], ignore_index=True)

factor_ts = (
    stocks_all.groupby("date")[["ret_1m","mom_6m","vol_12m","beta_3y",
                                 "mkt_ret_1m","mkt_vol_3m"]]
    .mean().dropna().sort_index()
)

# ── Derived stress features ───────────────────────────────────
factor_ts["vol_spike"]    = factor_ts["vol_12m"] / (
    factor_ts["vol_12m"].rolling(12, min_periods=3).mean() + 1e-9)
factor_ts["ret_drawdown"] = factor_ts["ret_1m"].rolling(3, min_periods=1).sum()
factor_ts["trend_str"]    = factor_ts["mom_6m"] / (factor_ts["vol_12m"] + 1e-6)
factor_ts[["vol_spike","ret_drawdown","trend_str"]] = (
    factor_ts[["vol_spike","ret_drawdown","trend_str"]].ffill().fillna(0)
)

# ── KMeans + GMM on full features ────────────────────────────
scaler_reg = RobustScaler()
X_factors  = scaler_reg.fit_transform(
    factor_ts[["ret_1m","mom_6m","vol_12m","beta_3y",
               "mkt_ret_1m","mkt_vol_3m"]].values)

km = KMeans(n_clusters=N_REGIMES, random_state=RANDOM_STATE, n_init=30)
factor_ts["regime_kmeans"] = km.fit_predict(X_factors)

gmm = GaussianMixture(n_components=N_REGIMES, covariance_type="full",
                      random_state=RANDOM_STATE, n_init=20)
factor_ts["regime_gmm"] = gmm.fit_predict(X_factors)

# ── HMM with smart initialisation ────────────────────────────
hmm_raw = factor_ts[[
    "ret_1m","vol_12m","vol_spike","ret_drawdown","trend_str",
]].values.astype(float)

hmm_scaler = StandardScaler()
hmm_input  = hmm_scaler.fit_transform(hmm_raw)

# Strategy: run with N_REGIMES+1=4, pick best, then merge closest pair
best_hmm, best_ll, best_n = None, -np.inf, N_REGIMES

for n_comp in [N_REGIMES, N_REGIMES+1]:
    for seed in range(30):
        try:
            h = GaussianHMM(
                n_components    = n_comp,
                n_iter          = 5000,
                covariance_type = "diag",
                min_covar       = 1e-3,
                random_state    = seed,
            )
            h.fit(hmm_input)
            ll = h.score(hmm_input)
            if ll > best_ll:
                best_ll, best_hmm, best_n = ll, h, n_comp
        except:
            pass

raw_labels = best_hmm.predict(hmm_input)

# If 4 components used, merge the two most similar
if best_n == N_REGIMES + 1:
    means = best_hmm.means_
    min_dist, merge_pair = np.inf, (0,1)
    for a in range(best_n):
        for b in range(a+1, best_n):
            d = np.linalg.norm(means[a] - means[b])
            if d < min_dist:
                min_dist, merge_pair = d, (a, b)
    label_map = {i: i for i in range(best_n)}
    label_map[merge_pair[1]] = merge_pair[0]
    raw_labels = np.array([label_map[l] for l in raw_labels])
    unique_l = sorted(set(raw_labels))
    reindex  = {old: new for new, old in enumerate(unique_l)}
    raw_labels = np.array([reindex[l] for l in raw_labels])

factor_ts["regime_hmm"] = raw_labels
factor_ts["regime"]     = factor_ts["regime_hmm"]

# ── Label by VOLATILITY ───────────────────────────────────────
vol_means   = factor_ts.groupby("regime")["vol_12m"].mean().sort_values()
ordered_ids = vol_means.index.tolist()
REGIME_NAMES = {
    ordered_ids[0]: "Low-Vol/Bull",
    ordered_ids[1]: "Transitional",
    ordered_ids[2]: "High-Vol/Stress",
}
factor_ts["regime_label"] = factor_ts["regime"].map(REGIME_NAMES)

# ── Validation ────────────────────────────────────────────────
print("\n=== REGIME VALIDATION ===")
check_dates = {
    "2008-10": "High-Vol/Stress",
    "2020-03": "High-Vol/Stress",
    "2022-06": "High-Vol/Stress",
}
for date_str, expected in check_dates.items():
    try:
        idx     = factor_ts.index.get_indexer([pd.Timestamp(date_str)], method="nearest")[0]
        closest = factor_ts.index[idx]
        actual  = factor_ts.loc[closest, "regime_label"]
        status  = "✅" if actual == expected else "⚠️"
        print(f"  {status} {date_str}: expected={expected}, got={actual}")
    except Exception as e:
        print(f"  ⚠️  {date_str}: {e}")

print(f"\nRegime distribution:\n{factor_ts['regime_label'].value_counts().to_string()}")
print(f"\nMean stats per regime:")
print(factor_ts.groupby("regime_label")[
    ["ret_1m","vol_12m","vol_spike","ret_drawdown"]
].mean().round(4).to_string())

# ── PCA plot (3 methods) ──────────────────────────────────────
pca   = PCA(n_components=2, random_state=RANDOM_STATE)
X_pca = pca.fit_transform(X_factors)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
for ax, (title, col) in zip(axes, [("K-Means","regime_kmeans"),
                                    ("GMM","regime_gmm"),
                                    ("HMM (Primary)","regime_hmm")]):
    for ri, c in zip(sorted(factor_ts[col].unique()), PALETTE):
        m = (factor_ts[col] == ri).values
        ax.scatter(X_pca[m,0], X_pca[m,1], c=c, s=20, alpha=0.8,
                   label=REGIME_NAMES.get(ri, f"R{ri}"))
    ax.set_title(title, fontsize=10)
    ax.set_xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)")
    ax.set_ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)")
    ax.legend(fontsize=7)
plt.suptitle("Regime Clustering — PCA 2D", fontsize=13, y=1.01)
plt.tight_layout(); plt.show()

# ── Regime Timeline ──────────────────────────────────────────
colors_t = {r: c for r, c in zip(sorted(factor_ts["regime"].unique()), PALETTE)}

fig, ax = plt.subplots(figsize=(14, 2.5))
periods = []
prev_r, start_d = None, None
for d, row in factor_ts.iterrows():
    r = int(row["regime"])
    if prev_r is None:
        prev_r, start_d = r, d
    elif r != prev_r:
        periods.append((start_d, d, prev_r))
        prev_r, start_d = r, d
if prev_r is not None:
    periods.append((start_d, factor_ts.index[-1], prev_r))

for start, end, r in periods:
    ax.axvspan(start, end, alpha=0.70,
               color=colors_t[r],
               label=REGIME_NAMES.get(r, f"R{r}"))

handles, labels = ax.get_legend_handles_labels()
unique = dict(zip(labels, handles))
ax.legend(unique.values(), unique.keys(), fontsize=9, loc="upper left")
ax.set_title("Market Regime Timeline (HMM) — up to today")
ax.set_xlabel("Date"); ax.set_yticks([])
ax.set_xlim(factor_ts.index[0], factor_ts.index[-1])
plt.tight_layout(); plt.show()

# Factor boxplots per regime
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, col in zip(axes, ["ret_1m","vol_12m","vol_spike","ret_drawdown"]):
    data_by_r = [
        factor_ts[factor_ts["regime"] == r][col].dropna().values
        for r in sorted(factor_ts["regime"].unique())
    ]
    bp = ax.boxplot(data_by_r, patch_artist=True)
    for patch, c in zip(bp["boxes"], PALETTE):
        patch.set_facecolor(c)
    ax.set_xticklabels(
        [REGIME_NAMES.get(r, f"R{r}") for r in sorted(factor_ts["regime"].unique())],
        fontsize=7)
    ax.set_title(col, fontsize=10)
plt.suptitle("Factor Distributions by Regime", fontsize=13)
plt.tight_layout(); plt.show()

# ============================================================
# SECTION 6 — MERGE REGIME & RE-SPLIT
# ============================================================
stocks_for_model = stocks_for_model.merge(
    factor_ts[["regime","regime_label"]],
    left_on="date", right_index=True, how="left"
)
stocks_for_model.dropna(subset=["regime"], inplace=True)
stocks_for_model["regime"] = stocks_for_model["regime"].astype(int)
stocks_for_model.reset_index(drop=True, inplace=True)

# Also merge regime into latest rows for signal generation
if len(stocks_latest) > 0:
    stocks_latest = stocks_latest.merge(
        factor_ts[["regime","regime_label"]],
        left_on="date", right_index=True, how="left"
    )
    # For any latest rows beyond factor_ts range, use the last known regime
    last_regime      = int(factor_ts["regime"].iloc[-1])
    last_regime_lbl  = REGIME_NAMES.get(last_regime, f"Regime {last_regime}")
    stocks_latest["regime"]       = stocks_latest["regime"].fillna(last_regime).astype(int)
    stocks_latest["regime_label"] = stocks_latest["regime_label"].fillna(last_regime_lbl)
    stocks_latest.reset_index(drop=True, inplace=True)

train_df = stocks_for_model[stocks_for_model["date"].isin(train_dates)].copy()
val_df   = stocks_for_model[stocks_for_model["date"].isin(val_dates)].copy()
test_df  = stocks_for_model[stocks_for_model["date"].isin(test_dates)].copy()
print(f"\n✅ Regime merged. Regimes: {sorted(stocks_for_model['regime'].unique())}")
if len(stocks_latest) > 0:
    print(f"   Latest date (signals): {stocks_latest['date'].max().date()} | "
          f"Regime: {stocks_latest[stocks_latest['date']==stocks_latest['date'].max()]['regime_label'].iloc[0]}")

# ============================================================
# SECTION 7 — OPTUNA-TUNED REGIME-SPECIFIC STACKING ENSEMBLES
# ============================================================
print("\n🤖 Tuning & training regime-specific stacking ensembles …")

def tune_xgb(X_tr, y_tr, X_vl, y_vl, n_trials=OPTUNA_TRIALS):
    def obj(trial):
        p = dict(
            n_estimators     = trial.suggest_int("n_estimators", 100, 500),
            max_depth        = trial.suggest_int("max_depth", 3, 7),
            learning_rate    = trial.suggest_float("learning_rate", 0.01, 0.15, log=True),
            subsample        = trial.suggest_float("subsample", 0.6, 1.0),
            colsample_bytree = trial.suggest_float("colsample_bytree", 0.5, 1.0),
            min_child_weight = trial.suggest_int("min_child_weight", 1, 10),
            reg_alpha        = trial.suggest_float("reg_alpha", 1e-4, 1.0, log=True),
            reg_lambda       = trial.suggest_float("reg_lambda", 1e-4, 5.0, log=True),
            eval_metric="logloss", random_state=RANDOM_STATE, verbosity=0,
        )
        m = XGBClassifier(**p)
        m.fit(X_tr, y_tr, eval_set=[(X_vl, y_vl)], verbose=False)
        prob = m.predict_proba(X_vl)[:,1]
        return roc_auc_score(y_vl, prob) if len(np.unique(y_vl)) > 1 else 0.5
    study = optuna.create_study(direction="maximize",
                                sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE))
    study.optimize(obj, n_trials=n_trials, show_progress_bar=False)
    return study.best_params

def tune_lgb(X_tr, y_tr, X_vl, y_vl, n_trials=OPTUNA_TRIALS):
    def obj(trial):
        p = dict(
            n_estimators     = trial.suggest_int("n_estimators", 100, 500),
            max_depth        = trial.suggest_int("max_depth", 3, 8),
            learning_rate    = trial.suggest_float("learning_rate", 0.01, 0.15, log=True),
            num_leaves       = trial.suggest_int("num_leaves", 20, 100),
            subsample        = trial.suggest_float("subsample", 0.6, 1.0),
            colsample_bytree = trial.suggest_float("colsample_bytree", 0.5, 1.0),
            min_child_samples= trial.suggest_int("min_child_samples", 5, 50),
            reg_alpha        = trial.suggest_float("reg_alpha", 1e-4, 1.0, log=True),
            random_state=RANDOM_STATE, verbosity=-1,
        )
        m = lgb.LGBMClassifier(**p)
        m.fit(X_tr, y_tr, eval_set=[(X_vl, y_vl)],
              callbacks=[lgb.early_stopping(20, verbose=False), lgb.log_evaluation(-1)])
        prob = m.predict_proba(X_vl)[:,1]
        return roc_auc_score(y_vl, prob) if len(np.unique(y_vl)) > 1 else 0.5
    study = optuna.create_study(direction="maximize",
                                sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE))
    study.optimize(obj, n_trials=n_trials, show_progress_bar=False)
    return study.best_params

def tune_rf(X_tr, y_tr, X_vl, y_vl, n_trials=OPTUNA_TRIALS//2):
    def obj(trial):
        p = dict(
            n_estimators     = trial.suggest_int("n_estimators", 100, 400),
            max_depth        = trial.suggest_int("max_depth", 3, 10),
            min_samples_leaf = trial.suggest_int("min_samples_leaf", 5, 30),
            max_features     = trial.suggest_float("max_features", 0.3, 1.0),
            random_state=RANDOM_STATE, n_jobs=-1,
        )
        m = RandomForestClassifier(**p)
        m.fit(X_tr, y_tr)
        prob = m.predict_proba(X_vl)[:,1]
        return roc_auc_score(y_vl, prob) if len(np.unique(y_vl)) > 1 else 0.5
    study = optuna.create_study(direction="maximize",
                                sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE))
    study.optimize(obj, n_trials=n_trials, show_progress_bar=False)
    return study.best_params

best_model_per_regime = {}
regime_val_perf       = {}
regime_cal_models     = {}

scaler_global = RobustScaler()
scaler_global.fit(train_df[FEATURE_COLS])

for regime_id in sorted(train_df["regime"].unique()):
    lbl = REGIME_NAMES.get(regime_id, f"Regime {regime_id}")
    print(f"\n── Regime {regime_id}: {lbl} ──")

    tr = train_df[train_df["regime"] == regime_id]
    vl = val_df[val_df["regime"] == regime_id]

    if len(tr) < 40 or len(vl) < 10:
        print(f"   ⚠️  Insufficient data (tr={len(tr)}, vl={len(vl)}) — global fallback")
        continue

    X_tr = scaler_global.transform(tr[FEATURE_COLS].values); y_tr = tr["target"].values
    X_vl = scaler_global.transform(vl[FEATURE_COLS].values); y_vl = vl["target"].values

    # 1. XGBoost
    print("   Tuning XGBoost …", end=" ")
    xgb_best_p = tune_xgb(X_tr, y_tr, X_vl, y_vl)
    xgb_m = XGBClassifier(
        **xgb_best_p,
        eval_metric="logloss", random_state=RANDOM_STATE, verbosity=0
    )
    xgb_m.fit(X_tr, y_tr, eval_set=[(X_vl, y_vl)], verbose=False)
    xgb_auc = roc_auc_score(y_vl, xgb_m.predict_proba(X_vl)[:,1]) if len(np.unique(y_vl))>1 else 0.5
    print(f"AUC={xgb_auc:.3f}")

    # 2. LightGBM
    print("   Tuning LightGBM …", end=" ")
    lgb_best_p = tune_lgb(X_tr, y_tr, X_vl, y_vl)
    lgb_m = lgb.LGBMClassifier(**lgb_best_p)
    lgb_m.fit(X_tr, y_tr, eval_set=[(X_vl, y_vl)],
              callbacks=[lgb.early_stopping(30, verbose=False), lgb.log_evaluation(-1)])
    lgb_auc = roc_auc_score(y_vl, lgb_m.predict_proba(X_vl)[:,1]) if len(np.unique(y_vl))>1 else 0.5
    print(f"AUC={lgb_auc:.3f}")

    # 3. RandomForest
    print("   Tuning RandomForest …", end=" ")
    rf_best_p = tune_rf(X_tr, y_tr, X_vl, y_vl)
    rf_m = RandomForestClassifier(**rf_best_p)
    rf_m.fit(X_tr, y_tr)
    rf_auc = roc_auc_score(y_vl, rf_m.predict_proba(X_vl)[:,1]) if len(np.unique(y_vl))>1 else 0.5
    print(f"AUC={rf_auc:.3f}")

    # 4. Stacking Ensemble
    print("   Building stacking ensemble …", end=" ")
    stack = StackingClassifier(
        estimators=[
            ("xgb", XGBClassifier(**xgb_best_p, eval_metric="logloss",
                                   random_state=RANDOM_STATE, verbosity=0)),
            ("lgb", lgb.LGBMClassifier(**lgb_best_p)),
            ("rf",  RandomForestClassifier(**rf_best_p)),
        ],
        final_estimator=LogisticRegression(C=1.0, max_iter=1000,
                                            random_state=RANDOM_STATE),
        cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE),
        stack_method="predict_proba", n_jobs=-1,
    )
    stack.fit(X_tr, y_tr)
    stack_auc = roc_auc_score(y_vl, stack.predict_proba(X_vl)[:,1]) if len(np.unique(y_vl))>1 else 0.5
    print(f"Stack AUC={stack_auc:.3f}")

    # 5. Pick best & calibrate
    best_base_auc = max(xgb_auc, lgb_auc, rf_auc, stack_auc)
    if   stack_auc == best_base_auc: raw_model, best_name = copy.deepcopy(stack), "Stacking Ensemble"
    elif xgb_auc   == best_base_auc: raw_model, best_name = copy.deepcopy(xgb_m), "XGBoost"
    elif lgb_auc   == best_base_auc: raw_model, best_name = copy.deepcopy(lgb_m), "LightGBM"
    else:                            raw_model, best_name = copy.deepcopy(rf_m),  "RandomForest"

    cal_model = CalibratedClassifierCV(raw_model, method="isotonic", cv="prefit")
    cal_model.fit(X_vl, y_vl)

    cal_prob = cal_model.predict_proba(X_vl)[:,1]
    cal_auc  = roc_auc_score(y_vl, cal_prob) if len(np.unique(y_vl))>1 else 0.5
    brier    = brier_score_loss(y_vl, cal_prob)
    print(f"   ✅ Best: {best_name} | Cal AUC={cal_auc:.3f} | Brier={brier:.3f}")

    best_model_per_regime[regime_id] = (best_name, cal_model)
    regime_cal_models[regime_id]     = cal_model
    regime_val_perf[regime_id] = {
        "AUC": cal_auc, "Brier": brier,
        "XGB_AUC": xgb_auc, "LGB_AUC": lgb_auc,
        "RF_AUC": rf_auc, "Stack_AUC": stack_auc,
    }

# ============================================================
# SECTION 8 — GLOBAL FALLBACK MODEL
# ============================================================
print("\n🌐 Training global LightGBM fallback …")
X_tr_g = scaler_global.transform(train_df[FEATURE_COLS].values)
y_tr_g = train_df["target"].values
X_vl_g = scaler_global.transform(val_df[FEATURE_COLS].values)
y_vl_g = val_df["target"].values

global_lgb = lgb.LGBMClassifier(n_estimators=400, max_depth=5, learning_rate=0.04,
                                  num_leaves=50, random_state=RANDOM_STATE, verbosity=-1)
global_lgb.fit(X_tr_g, y_tr_g, eval_set=[(X_vl_g, y_vl_g)],
               callbacks=[lgb.early_stopping(30, verbose=False), lgb.log_evaluation(-1)])
global_cal = CalibratedClassifierCV(global_lgb, method="isotonic", cv="prefit")
global_cal.fit(X_vl_g, y_vl_g)
g_auc = roc_auc_score(y_vl_g, global_cal.predict_proba(X_vl_g)[:,1])
print(f"   Global Cal AUC: {g_auc:.3f}")

for rid in sorted(stocks_for_model["regime"].unique()):
    if rid not in best_model_per_regime:
        best_model_per_regime[rid] = ("LightGBM (global fallback)", global_cal)
        print(f"   ℹ️  Regime {rid}: global fallback assigned.")

# ============================================================
# SECTION 9 — PERFORMANCE PLOTS
# ============================================================
print("\n📊 Plotting validation performance …")

if regime_val_perf:
    fig, axes = plt.subplots(1, len(regime_val_perf), figsize=(6*len(regime_val_perf), 4))
    if len(regime_val_perf) == 1: axes = [axes]
    for ax, (rid, perf) in zip(axes, regime_val_perf.items()):
        models = ["XGB","LGB","RF","Stack","Calibrated"]
        aucs   = [perf.get("XGB_AUC",0), perf.get("LGB_AUC",0),
                  perf.get("RF_AUC",0),  perf.get("Stack_AUC",0), perf.get("AUC",0)]
        colors = ["#2196F3","#4CAF50","#FF9800","#9C27B0","#E91E63"]
        bars   = ax.bar(models, aucs, color=colors, edgecolor="white", linewidth=1.2)
        ax.set_ylim(0.4, 0.85)
        ax.axhline(0.5, color="red", lw=1, ls="--", label="Random")
        for b, v in zip(bars, aucs):
            ax.text(b.get_x()+b.get_width()/2, v+0.005, f"{v:.3f}",
                    ha="center", va="bottom", fontsize=8)
        lbl = REGIME_NAMES.get(rid, f"R{rid}")
        ax.set_title(f"{lbl}\n({best_model_per_regime[rid][0]})", fontsize=9)
        ax.set_ylabel("Val AUC"); ax.legend(fontsize=8)
    plt.suptitle("Model Comparison by Regime (Validation AUC)", fontsize=13)
    plt.tight_layout(); plt.show()

# Confusion matrices
n_r = len(best_model_per_regime)
fig, axes = plt.subplots(1, max(n_r,1), figsize=(5*max(n_r,1), 4))
if n_r == 1: axes = [axes]
for ax, (rid, (mname, model)) in zip(axes, best_model_per_regime.items()):
    te = test_df[test_df["regime"] == rid]
    if len(te) < 5: ax.set_visible(False); continue
    X_te = scaler_global.transform(te[FEATURE_COLS].values); y_te = te["target"].values
    y_pred = model.predict(X_te)
    ConfusionMatrixDisplay(confusion_matrix(y_te, y_pred),
                           display_labels=["Underperf","Outperf"]).plot(
        ax=ax, colorbar=False, cmap="Blues")
    ax.set_title(f"{REGIME_NAMES.get(rid, f'R{rid}')}\n({mname})", fontsize=8)
plt.suptitle("Confusion Matrices — Test Set", fontsize=11)
plt.tight_layout(); plt.show()

# ROC curves
fig, ax = plt.subplots(figsize=(7, 5))
for (rid, (mname, model)), c in zip(best_model_per_regime.items(), PALETTE):
    te = test_df[test_df["regime"] == rid]
    if len(te) < 10: continue
    X_te = scaler_global.transform(te[FEATURE_COLS].values); y_te = te["target"].values
    if len(np.unique(y_te)) < 2: continue
    y_prob = model.predict_proba(X_te)[:,1]
    auc_   = roc_auc_score(y_te, y_prob)
    RocCurveDisplay.from_predictions(y_te, y_prob, ax=ax, color=c,
        name=f"{REGIME_NAMES.get(rid,'R'+str(rid))} AUC={auc_:.3f}")
ax.plot([0,1],[0,1],"k--",lw=0.8)
ax.set_title("ROC Curves per Regime (Test)")
plt.tight_layout(); plt.show()

# Feature importance
n_rp = len(best_model_per_regime)
fig, axes = plt.subplots(1, max(n_rp,1), figsize=(7*max(n_rp,1), 5))
if n_rp == 1: axes = [axes]
for ax, (rid, (mname, model)) in zip(axes, best_model_per_regime.items()):
    lbl = REGIME_NAMES.get(rid, f"R{rid}")
    try:
        base = model.calibrated_classifiers_[0].estimator
        if hasattr(base, "feature_importances_"):
            imp = base.feature_importances_
        elif hasattr(base, "estimators_"):
            imp = base.estimators_[0].feature_importances_
        else:
            imp = None
        if imp is not None:
            idx = np.argsort(imp)[::-1][:12]
            ax.barh([FEATURE_COLS[i] for i in idx][::-1], imp[idx][::-1], color=PALETTE[0])
            ax.set_title(f"{lbl}\n({mname})", fontsize=9)
            ax.set_xlabel("Importance")
    except Exception as e:
        ax.set_title(f"{lbl} — N/A")
plt.suptitle("Feature Importance by Regime", fontsize=13)
plt.tight_layout(); plt.show()
print("   ✅ Model evaluation plots done.")

# ============================================================
# SECTION 10 — WALK-FORWARD PORTFOLIO SIMULATION
# ============================================================
print("\n💼 Walk-forward portfolio simulation …")

nifty_ret_lookup = nifty_idx.set_index("date")["nifty_ret"]

def get_regime(date):
    if date in factor_ts.index:
        return int(factor_ts.loc[date, "regime"])
    i = min(factor_ts.index.searchsorted(date), len(factor_ts)-1)
    return int(factor_ts.iloc[i]["regime"])

def build_portfolio(today_df, regime_id):
    mname, model = best_model_per_regime.get(regime_id, ("global", global_cal))
    today = today_df.dropna(subset=FEATURE_COLS).copy()
    if len(today) < TOP_K: return None, None, None
    X = scaler_global.transform(today[FEATURE_COLS].values)
    today["score"] = model.predict_proba(X)[:,1]
    top = today.nlargest(TOP_K, "score").copy()
    vol = top["vol_12m"].replace(0, np.nan).fillna(0.04).clip(lower=1e-6)
    w   = 1.0/vol; w /= w.sum()
    w   = w.clip(upper=MAX_WEIGHT); w /= w.sum()
    top["sec_"] = top["stock"].map(SECTOR_MAP).fillna("Other")
    for _, grp in top.groupby("sec_"):
        sw = w.loc[grp.index].sum()
        if sw > SECTOR_CAP: w.loc[grp.index] *= SECTOR_CAP/sw
    w /= w.sum()
    return top["stock"].values, w.values, mname

sim_dates = sorted(test_dates)
rec_regime, rec_global_r, rec_equal, rec_nifty = [], [], [], []
prev_w_r = prev_w_g = None
monthly_picks_log = []

for i in range(len(sim_dates)-1):
    today_date = sim_dates[i]; next_date = sim_dates[i+1]
    today_df   = test_df[test_df["date"] == today_date].copy()
    next_df    = test_df[test_df["date"] == next_date].set_index("stock")
    nifty_nxt  = float(nifty_ret_lookup.get(next_date, 0.0))
    if len(today_df) < TOP_K: continue

    hist_ser = nifty_ret_lookup.loc[nifty_ret_lookup.index <= today_date]
    hist_vol = float(hist_ser.rolling(12).std().iloc[-1]) if len(hist_ser)>=12 else 0.04
    exposure = float(np.clip(TARGET_VOL/(max(hist_vol,1e-4)*np.sqrt(12)), 0.5, 1.3))
    rid      = get_regime(today_date)

    sel, wts, _ = build_portfolio(today_df, rid)
    if sel is not None:
        realized = np.array([float(next_df["ret_1m"].get(s, 0.0)) for s in sel])
        turnover = (float(np.sum(np.abs(wts - prev_w_r)))
                    if prev_w_r is not None and len(prev_w_r)==len(wts) else 1.0)
        prev_w_r = wts.copy()
        net = exposure*float(np.dot(wts, realized)) - SLIPPAGE - IMPACT*turnover
        rec_regime.append((next_date, net))
        monthly_picks_log.append({
            "date": today_date, "regime": rid,
            "regime_label": REGIME_NAMES.get(rid,"?"),
            "stocks": list(sel), "weights": list(wts), "net_ret": net,
        })

    tg = today_df.dropna(subset=FEATURE_COLS).copy()
    if len(tg) >= TOP_K:
        Xg = scaler_global.transform(tg[FEATURE_COLS].values)
        tg["sg"] = global_cal.predict_proba(Xg)[:,1]
        top_g = tg.nlargest(TOP_K, "sg")
        vg    = top_g["vol_12m"].replace(0,np.nan).fillna(0.04).clip(lower=1e-6)
        wg    = 1.0/vg; wg /= wg.sum(); wg = wg.clip(upper=MAX_WEIGHT); wg /= wg.sum()
        rg    = np.array([float(next_df["ret_1m"].get(s,0.0)) for s in top_g["stock"]])
        tg_   = float(np.sum(np.abs(wg.values - prev_w_g))) if prev_w_g is not None else 1.0
        prev_w_g = wg.values.copy()
        net_g = exposure*float(np.dot(wg.values, rg)) - SLIPPAGE - IMPACT*tg_
        rec_global_r.append((next_date, net_g))

    ew = float(next_df["ret_1m"].mean()) if len(next_df)>0 else 0.0
    rec_equal.append((next_date, ew))
    rec_nifty.append((next_date, nifty_nxt))

def to_series(lst):
    d = pd.DataFrame(lst, columns=["date","ret"])
    d["date"] = pd.to_datetime(d["date"])
    return d.set_index("date")["ret"]

r_regime = to_series(rec_regime)
r_global = to_series(rec_global_r)
r_equal  = to_series(rec_equal)
r_nifty  = to_series(rec_nifty)

common_idx = (r_regime.index.intersection(r_global.index).intersection(r_nifty.index))
r_regime = r_regime.reindex(common_idx)
r_global = r_global.reindex(common_idx)
r_equal  = r_equal.reindex(common_idx).fillna(0)
r_nifty  = r_nifty.reindex(common_idx)
print(f"   Simulation months: {len(common_idx)}")

# ============================================================
# SECTION 11 — HYPOTHETICAL PORTFOLIO TRACKER
# ============================================================
print("\n📈 Computing hypothetical ₹10L portfolio …")

eq_regime_val = INITIAL_CAPITAL * (1 + r_regime).cumprod()
eq_global_val = INITIAL_CAPITAL * (1 + r_global).cumprod()
eq_equal_val  = INITIAL_CAPITAL * (1 + r_equal).cumprod()
eq_nifty_val  = INITIAL_CAPITAL * (1 + r_nifty).cumprod()

eq_regime = (1+r_regime).cumprod()
eq_global = (1+r_global).cumprod()
eq_equal  = (1+r_equal).cumprod()
eq_nifty  = (1+r_nifty).cumprod()

def drawdown_series(s):
    eq_ = (1+s).cumprod()
    return (eq_ - eq_.cummax()) / eq_.cummax()

def risk_report(s):
    n      = len(s)
    cagr   = (1+s).prod()**(12/n)-1
    ann_v  = s.std()*np.sqrt(12)
    sharpe = (s.mean()/s.std())*np.sqrt(12) if s.std()>0 else np.nan
    eq_    = (1+s).cumprod(); dd_ = (eq_-eq_.cummax())/eq_.cummax()
    mdd    = float(dd_.min())
    calmar = cagr/abs(mdd) if mdd!=0 else np.nan
    hit    = float((s>0).mean())
    neg_s  = s[s<0]
    sor_d  = neg_s.std()*np.sqrt(12) if len(neg_s)>1 else np.nan
    sort_  = (s.mean()*12)/sor_d if (sor_d and sor_d>0) else np.nan
    aln    = pd.concat([s, r_nifty], axis=1).dropna(); aln.columns = ["s","n"]
    if len(aln)>5 and aln["n"].std()>0:
        cv   = np.cov(aln["s"].values, aln["n"].values)
        beta = cv[0,1]/cv[1,1]
        alph = (aln["s"].mean()-beta*aln["n"].mean())*12
    else:
        beta, alph = np.nan, np.nan
    final = INITIAL_CAPITAL*(1+s).prod()
    return {
        "CAGR (%)":      round(cagr*100, 2),
        "Ann Vol (%)":   round(ann_v*100, 2),
        "Sharpe":        round(sharpe, 3)  if not np.isnan(sharpe) else "N/A",
        "Sortino":       round(sort_, 3)   if not np.isnan(sort_)  else "N/A",
        "Max DD (%)":    round(mdd*100, 2),
        "Calmar":        round(calmar, 3)  if not np.isnan(calmar) else "N/A",
        "Hit Rate (%)":  round(hit*100, 2),
        "Beta":          round(beta, 3)    if not np.isnan(beta)   else "N/A",
        "Alpha ann (%)": round(alph*100,2) if not np.isnan(alph)   else "N/A",
        "Final Value ₹": f"₹{final:,.0f}",
    }

report = pd.DataFrame({
    "Regime-Aware ML": risk_report(r_regime),
    "Global ML":       risk_report(r_global),
    "Equal-Weight":    risk_report(r_equal),
    "NIFTY 50":        risk_report(r_nifty),
}).T

print("\n" + "="*72)
print("            COMPREHENSIVE RISK & PERFORMANCE REPORT")
print("="*72)
print(report.to_string())
print("="*72)

# Portfolio value chart
fig, axes = plt.subplots(2, 1, figsize=(14,10))
ax = axes[0]
ax.plot(eq_regime_val/1e5, lw=2.5, color="#2196F3", label="Regime-Aware ML (Our Model)")
ax.plot(eq_global_val/1e5, lw=1.8, color="#FF9800", ls="--", label="Global ML")
ax.plot(eq_equal_val/1e5,  lw=1.5, color="#9C27B0", ls="-.", label="Equal-Weight")
ax.plot(eq_nifty_val/1e5,  lw=2,   color="#4CAF50", label="NIFTY 50 Benchmark")
ax.axhline(INITIAL_CAPITAL/1e5, color="gray", lw=1, ls=":", label="Initial ₹10L")
ax.set_title("₹10,00,000 Invested at Test Period Start", fontsize=13, fontweight="bold")
ax.set_ylabel("Portfolio Value (₹ Lakhs)"); ax.set_xlabel("Date")
ax.legend(fontsize=10); ax.grid(alpha=0.3)
for series, color in [(eq_regime_val,"#2196F3"),(eq_nifty_val,"#4CAF50")]:
    fv = series.iloc[-1]
    ax.annotate(f"₹{fv/1e5:.1f}L",
                xy=(series.index[-1], fv/1e5),
                xytext=(10,0), textcoords="offset points",
                fontsize=9, color=color, fontweight="bold")
ax2 = axes[1]
ax2.fill_between(r_regime.index, drawdown_series(r_regime)*100, 0,
                  alpha=0.6, color="#2196F3", label="Regime-Aware ML")
ax2.fill_between(r_nifty.index,  drawdown_series(r_nifty)*100,  0,
                  alpha=0.4, color="#4CAF50",  label="NIFTY 50")
ax2.set_title("Drawdown (%)", fontsize=12)
ax2.set_ylabel("Drawdown %"); ax2.legend(); ax2.grid(alpha=0.3)
plt.tight_layout(); plt.show()

# Year-wise returns
def yearly_returns(s, label):
    d = pd.DataFrame({"date":s.index,"ret":s.values})
    d["year"] = pd.to_datetime(d["date"]).dt.year
    return d.groupby("year")["ret"].apply(lambda x:(1+x).prod()-1).rename(label)

yr_all = pd.concat([
    yearly_returns(r_regime,"Regime-Aware ML"),
    yearly_returns(r_global,"Global ML"),
    yearly_returns(r_equal, "Equal-Weight"),
    yearly_returns(r_nifty, "NIFTY 50"),
],axis=1).dropna()

x_pos = np.arange(len(yr_all)); w_bar = 0.2
fig, ax = plt.subplots(figsize=(13, 5))
ax.bar(x_pos-1.5*w_bar, yr_all["Regime-Aware ML"]*100, w_bar, label="Regime-Aware ML", color="#2196F3")
ax.bar(x_pos-0.5*w_bar, yr_all["Global ML"]*100,        w_bar, label="Global ML",        color="#FF9800")
ax.bar(x_pos+0.5*w_bar, yr_all["Equal-Weight"]*100,     w_bar, label="Equal-Weight",     color="#9C27B0")
ax.bar(x_pos+1.5*w_bar, yr_all["NIFTY 50"]*100,         w_bar, label="NIFTY 50",         color="#4CAF50")
ax.set_xticks(x_pos); ax.set_xticklabels(yr_all.index)
ax.axhline(0, color="black", lw=0.8)
ax.set_title("Year-wise Returns Comparison (%)", fontsize=13)
ax.set_ylabel("Annual Return (%)"); ax.legend(); ax.grid(alpha=0.3, axis="y")
plt.tight_layout(); plt.show()
print("\n===== YEAR-WISE RETURNS (%) =====")
print((yr_all*100).round(2).to_string())

def rolling_annual(s, w=12):
    return s.rolling(w).apply(lambda x:(1+x).prod()-1, raw=True)

fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(rolling_annual(r_regime), color="#2196F3", lw=2, label="Regime-Aware ML")
ax.plot(rolling_annual(r_global), color="#FF9800", ls="--", label="Global ML")
ax.plot(rolling_annual(r_nifty),  color="#4CAF50", lw=2, label="NIFTY 50")
ax.axhline(0, color="black", lw=0.8)
ax.set_title("Rolling 12-Month Return", fontsize=13)
ax.set_ylabel("Return"); ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

# ============================================================
# SECTION 12 — BUY / HOLD / SELL RATINGS  (truly latest data)
# ============================================================
print("\n🎯 Generating BUY / HOLD / SELL ratings …")

def rating(p):
    if   p >= BUY_THRESH:  return "🟢 BUY"
    elif p <= SELL_THRESH: return "🔴 SELL"
    else:                  return "🟡 HOLD"

def conviction(p):
    if   p >= 0.72: return "★★★ HIGH"
    elif p >= 0.62: return "★★☆ MEDIUM"
    elif p >= 0.50: return "★☆☆ LOW"
    elif p >= 0.42: return "☆☆☆ HOLD"
    else:           return "▼▼▼ EXIT"

# ── Choose signal source ─────────────────────────────────────
# Priority: stocks_latest (truly current, no forward leakage)
#           Fallback: most recent date in stocks_for_model
if len(stocks_latest) > 0:
    signal_date  = stocks_latest["date"].max()
    signal_df    = stocks_latest[stocks_latest["date"] == signal_date].copy()
    signal_src   = "LIVE (no forward returns yet)"
else:
    signal_date  = stocks_for_model["date"].max()
    signal_df    = stocks_for_model[stocks_for_model["date"] == signal_date].copy()
    signal_src   = "most recent complete data"

latest_regime = int(signal_df["regime"].mode().iloc[0])
latest_label  = REGIME_NAMES.get(latest_regime, f"Regime {latest_regime}")

_, cur_model  = best_model_per_regime[latest_regime]
cur_clean     = signal_df.dropna(subset=FEATURE_COLS).copy()
X_cur         = scaler_global.transform(cur_clean[FEATURE_COLS].values)
cur_clean["prob"]       = cur_model.predict_proba(X_cur)[:,1]
cur_clean["rating"]     = cur_clean["prob"].apply(rating)
cur_clean["conviction"] = cur_clean["prob"].apply(conviction)
cur_clean["prob_pct"]   = (cur_clean["prob"]*100).round(1)

ratings_full = (cur_clean
    .sort_values("prob", ascending=False)
    [["stock","sector","rating","conviction","prob_pct",
      "mom_6m","vol_12m","rsi_14","beta_3y"]]
    .reset_index(drop=True))

print(f"\n   Signal Source: {signal_src}")
print(f"   Date: {signal_date.date()}  |  Regime: {latest_label}  |  "
      f"Model: {best_model_per_regime[latest_regime][0]}")
print(f"   BUY ≥{BUY_THRESH*100:.0f}%  |  SELL ≤{SELL_THRESH*100:.0f}%\n")
print(f"{'='*90}")
print(f"{'#':<3} {'Stock':<15} {'Sector':<12} {'Rating':<12} {'Conviction':<14} "
      f"{'Prob%':<7} {'Mom6M':>7} {'Vol12M':>7} {'RSI':>6} {'Beta':>6}")
print(f"{'='*90}")
for i, row in ratings_full.iterrows():
    print(f"{i+1:<3} {row['stock']:<15} {row['sector']:<12} {row['rating']:<12} "
          f"{row['conviction']:<14} {row['prob_pct']:<7.1f} "
          f"{row['mom_6m']*100:>6.1f}% {row['vol_12m']*100:>6.1f}% "
          f"{row['rsi_14']:>6.1f} {row['beta_3y']:>6.2f}")
print(f"{'='*90}")

buys  = ratings_full[ratings_full["rating"] == "🟢 BUY"]
holds = ratings_full[ratings_full["rating"] == "🟡 HOLD"]
sells = ratings_full[ratings_full["rating"] == "🔴 SELL"]
print(f"\n   BUY : {len(buys)}  |  HOLD : {len(holds)}  |  SELL : {len(sells)}")

# Rating chart
fig, axes = plt.subplots(1, 2, figsize=(16,6))
ax = axes[0]
colors_map = {"🟢 BUY":"#4CAF50","🟡 HOLD":"#FF9800","🔴 SELL":"#F44336"}
bar_colors = [colors_map[r] for r in ratings_full["rating"]]
bars = ax.barh(ratings_full["stock"][::-1], ratings_full["prob_pct"][::-1],
               color=bar_colors[::-1], edgecolor="white", linewidth=0.5)
ax.axvline(BUY_THRESH*100,  color="#4CAF50", lw=1.5, ls="--", label=f"BUY ≥{BUY_THRESH*100:.0f}%")
ax.axvline(SELL_THRESH*100, color="#F44336", lw=1.5, ls="--", label=f"SELL ≤{SELL_THRESH*100:.0f}%")
ax.axvline(50, color="gray", lw=1, ls=":", label="50% random")
ax.set_xlabel("Outperform Probability (%)")
ax.set_title(f"BUY / HOLD / SELL — All Stocks\n{signal_date.date()} | {latest_label} | {signal_src}", fontsize=10)
ax.legend(fontsize=8); ax.set_xlim(30, 80)
for bar, val in zip(bars[::-1], ratings_full["prob_pct"]):
    ax.text(val+0.3, bar.get_y()+bar.get_height()/2, f"{val:.1f}%",
            va="center", fontsize=7)

ax2 = axes[1]
sector_buy = buys["sector"].value_counts()
if len(sector_buy) > 0:
    wedge_colors = plt.cm.Set3(np.linspace(0,1,len(sector_buy)))
    ax2.pie(sector_buy.values, labels=sector_buy.index, autopct="%1.0f%%",
            colors=wedge_colors, startangle=90, textprops={"fontsize":9})
    ax2.set_title(f"BUY Picks by Sector\n({len(buys)} BUY stocks)", fontsize=11)
else:
    ax2.text(0.5,0.5,"No BUY signals",ha="center",va="center")
    ax2.set_title("BUY Picks by Sector")
plt.tight_layout(); plt.show()

print(f"\n{'='*60}")
print(f"   🟢 TOP BUY PICKS — {latest_label.upper()}")
print(f"   📅 As of {signal_date.date()} | Horizon: 6 months")
print(f"   ⚡ Source: {signal_src}")
print(f"{'='*60}")
for i, row in buys.head(10).iterrows():
    print(f"  {row['stock']:<16} | {row['conviction']:<14} | "
          f"Prob={row['prob_pct']:.1f}% | Mom6M={row['mom_6m']*100:.1f}% | RSI={row['rsi_14']:.0f}")

if len(sells) > 0:
    print(f"\n{'='*60}")
    print(f"   🔴 SELL / AVOID SIGNALS")
    print(f"{'='*60}")
    for i, row in sells.iterrows():
        print(f"  {row['stock']:<16} | Prob={row['prob_pct']:.1f}% | Mom6M={row['mom_6m']*100:.1f}%")

print("\n\n===== RECOMMENDATIONS BY REGIME — FORWARD OUTLOOK =====")
for rid, (mname, model) in best_model_per_regime.items():
    # Use stocks_latest rows for that regime if available, else stocks_for_model
    if len(stocks_latest) > 0:
        rdf   = stocks_latest[stocks_latest["regime"] == rid].copy()
    else:
        rdf   = stocks_for_model[stocks_for_model["regime"] == rid].copy()
    if len(rdf) == 0:
        rdf = stocks_for_model[stocks_for_model["regime"] == rid].copy()
    r_lat = rdf["date"].max()
    sub   = rdf[rdf["date"] == r_lat].dropna(subset=FEATURE_COLS).copy()
    if len(sub) < TOP_K: continue
    X_ = scaler_global.transform(sub[FEATURE_COLS].values)
    sub["prob"]     = model.predict_proba(X_)[:,1]
    sub["rating"]   = sub["prob"].apply(rating)
    sub["prob_pct"] = (sub["prob"]*100).round(1)
    lbl_ = REGIME_NAMES.get(rid, f"R{rid}")
    print(f"\n--- {lbl_} ({mname}) | as of {r_lat.date()} ---")
    picks = sub.sort_values("prob", ascending=False).head(TOP_K)[
        ["stock","sector","rating","prob_pct","mom_6m"]].reset_index(drop=True)
    picks["mom_6m"] = (picks["mom_6m"]*100).round(1).astype(str)+"%"
    print(picks.to_string(index=False))

# ============================================================
# SECTION 13 — FINAL DASHBOARD
# ============================================================
print("\n📊 Final dashboard …")

fig = plt.figure(figsize=(20, 16))
gs  = gridspec.GridSpec(3, 3, figure=fig, hspace=0.55, wspace=0.40)

ax1 = fig.add_subplot(gs[0,:2])
ax1.plot(eq_regime_val/1e5, lw=2.5, color="#2196F3", label="Regime-Aware ML")
ax1.plot(eq_global_val/1e5, lw=1.8, color="#FF9800", ls="--", label="Global ML")
ax1.plot(eq_nifty_val/1e5,  lw=2,   color="#4CAF50", label="NIFTY 50")
ax1.axhline(INITIAL_CAPITAL/1e5, color="gray", lw=1, ls=":", label="Initial ₹10L")
ax1.set_title("Portfolio Value (₹ Lakhs) — Test Period", fontsize=11)
ax1.set_ylabel("₹ Lakhs"); ax1.legend(fontsize=8); ax1.grid(alpha=0.3)

ax2 = fig.add_subplot(gs[0,2])
rating_counts = ratings_full["rating"].value_counts()
rc_clean = {k.split(" ")[-1]:v for k,v in rating_counts.items()}
ax2.pie([rc_clean.get("BUY",0), rc_clean.get("HOLD",0), rc_clean.get("SELL",0)],
        labels=[f"BUY ({rc_clean.get('BUY',0)})",
                f"HOLD ({rc_clean.get('HOLD',0)})",
                f"SELL ({rc_clean.get('SELL',0)})"],
        autopct="%1.0f%%",
        colors=["#4CAF50","#FF9800","#F44336"],
        startangle=90, wedgeprops=dict(width=0.5))
ax2.set_title(f"Signal Distribution\n{signal_date.date()}", fontsize=10)

ax3 = fig.add_subplot(gs[1,:2])
ax3.fill_between(r_regime.index, drawdown_series(r_regime)*100, 0,
                  alpha=0.6, color="#2196F3", label="Regime-Aware ML")
ax3.fill_between(r_nifty.index,  drawdown_series(r_nifty)*100,  0,
                  alpha=0.4, color="#4CAF50",  label="NIFTY 50")
ax3.set_title("Drawdown (%)", fontsize=11)
ax3.set_ylabel("%"); ax3.legend(); ax3.grid(alpha=0.3)

ax4 = fig.add_subplot(gs[1,2])
yr_plot = yr_all[["Regime-Aware ML","NIFTY 50"]]*100
xp_ = np.arange(len(yr_plot)); wb = 0.35
ax4.bar(xp_-wb/2, yr_plot["Regime-Aware ML"], wb, color="#2196F3", label="Regime ML")
ax4.bar(xp_+wb/2, yr_plot["NIFTY 50"],        wb, color="#4CAF50", label="NIFTY 50")
ax4.set_xticks(xp_); ax4.set_xticklabels(yr_plot.index, rotation=45, fontsize=7)
ax4.axhline(0, color="black"); ax4.set_title("Year-wise Returns (%)", fontsize=10)
ax4.legend(fontsize=7)

ax5 = fig.add_subplot(gs[2,:])
ax5.axis("off")
dcols = ["CAGR (%)","Ann Vol (%)","Sharpe","Sortino","Max DD (%)","Calmar",
         "Hit Rate (%)","Alpha ann (%)","Final Value ₹"]
tbl_data = report[dcols].reset_index()
tbl = ax5.table(cellText=tbl_data.values.tolist(),
                colLabels=["Strategy"]+dcols,
                cellLoc="center", loc="center")
tbl.auto_set_font_size(False); tbl.set_fontsize(8); tbl.scale(1, 2.2)
for j in range(len(dcols)+1):
    tbl[0,j].set_facecolor("#1565C0")
    tbl[0,j].set_text_props(color="white", fontweight="bold")
for i in range(1, 5):
    tbl[i,0].set_facecolor("#E3F2FD")
ax5.set_title("Performance Summary", pad=16, fontsize=11, fontweight="bold")

plt.suptitle("ENHANCED ADAPTIVE FACTOR-REGIME ML — NIFTY 50\nFull Pipeline Dashboard",
             fontsize=14, fontweight="bold", y=1.01)
plt.show()
print("   ✅ Final dashboard plotted.")

# ============================================================
# SECTION 14 — REGIME-WISE EQUITY CURVES
# ============================================================
n_rp = len(best_model_per_regime)
fig, axes = plt.subplots(1, max(n_rp,1), figsize=(6*max(n_rp,1), 5))
if n_rp == 1: axes = [axes]
for ax, ((rid, _), c) in zip(axes, zip(best_model_per_regime.items(), PALETTE)):
    r_dates = set(test_df[test_df["regime"]==rid]["date"].unique())
    sr = r_regime[r_regime.index.isin(r_dates)]
    sn = r_nifty[r_nifty.index.isin(r_dates)]
    if len(sr) < 2: ax.set_visible(False); continue
    ax.plot((1+sr).cumprod().values, lw=2, color=c, label="Regime-Aware ML")
    ax.plot((1+sn).cumprod().values, lw=2, color="#4CAF50", ls="--", label="NIFTY 50")
    lbl_ = REGIME_NAMES.get(int(rid), f"R{rid}")
    ax.set_title(f"{lbl_}\n({best_model_per_regime[rid][0]})", fontsize=9)
    ax.set_xlabel("Months"); ax.set_ylabel("Cum. Return")
    ax.legend(fontsize=8); ax.grid(alpha=0.3)
plt.suptitle("Regime-wise Equity Curves (ML vs NIFTY)", fontsize=13)
plt.tight_layout(); plt.show()

# ============================================================
# SECTION 15 — CLASSIFICATION REPORT (TEST)
# ============================================================
print("\n📋 Classification Report — TEST set …")
all_y, all_yp_r, all_yp_g = [], [], []
for rid, (mname, model) in best_model_per_regime.items():
    te = test_df[test_df["regime"] == rid]
    if len(te) < 5: continue
    X_te  = scaler_global.transform(te[FEATURE_COLS].values); y_te = te["target"].values
    yp_r  = model.predict(X_te)
    yp_g  = global_cal.predict(scaler_global.transform(te[FEATURE_COLS].values))
    print(f"\n--- {REGIME_NAMES.get(rid, f'R{rid}')} ({mname}) ---")
    print(classification_report(y_te, yp_r, target_names=["Underperf","Outperf"]))
    all_y.extend(y_te.tolist()); all_yp_r.extend(yp_r.tolist()); all_yp_g.extend(yp_g.tolist())

if all_y:
    print("\n===== OVERALL: Regime-Aware vs Global =====")
    print(f"{'Metric':<12} {'Regime-Aware':>14} {'Global ML':>12}")
    print("-"*40)
    for met, fn in [
        ("Precision", lambda y,p: precision_score(y,p, zero_division=0)),
        ("Recall",    lambda y,p: recall_score(y,p,    zero_division=0)),
        ("F1",        lambda y,p: f1_score(y,p,        zero_division=0)),
        ("AUC",       lambda y,p: roc_auc_score(y,p) if len(np.unique(y))>1 else 0.5),
    ]:
        try:
            rv = fn(all_y, all_yp_r); gv = fn(all_y, all_yp_g)
            print(f"{met:<12} {rv:>14.4f} {gv:>12.4f}")
        except: pass

# ============================================================
# FINAL SUMMARY
# ============================================================
print("\n"+"="*72)
print("   ✅  ENHANCED ADAPTIVE FACTOR-REGIME ML PIPELINE — COMPLETE")
print("="*72)
print(f"   Features engineered : {len(FEATURE_COLS)}")
print(f"   Regimes detected    : {N_REGIMES} (K-Means + GMM + HMM, best of 60 seeds)")
print(f"   Models per regime   : XGBoost + LightGBM + RF + Stacking + Isotonic Calib.")
print(f"   Stocks in universe  : {stocks_for_model['stock'].nunique()}")
print(f"   Simulation period   : {common_idx[0].date()} → {common_idx[-1].date()}")
print(f"   Signal date (live)  : {signal_date.date()} ({signal_src})")
try:
    print(f"   Regime-Aware CAGR   : {report.loc['Regime-Aware ML','CAGR (%)']:.2f}%")
    print(f"   NIFTY 50 CAGR       : {report.loc['NIFTY 50','CAGR (%)']:.2f}%")
    print(f"   Sharpe Ratio        : {report.loc['Regime-Aware ML','Sharpe']}")
    print(f"   Final Portfolio Val : {report.loc['Regime-Aware ML','Final Value ₹']}")
except: pass
print(f"\n   CURRENT SIGNALS ({signal_date.date()}):")
print(f"   Regime  : {latest_label}")
print(f"   BUY     : {', '.join(buys.head(5)['stock'].tolist())}")
if len(sells) > 0:
    print(f"   SELL    : {', '.join(sells['stock'].tolist())}")
print("="*72)
```
